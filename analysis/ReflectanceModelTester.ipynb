```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
from astropy.constants import c
c = c.value*1e-6
```

```python
nu = np.r_[140:160:1000j]
R = 0.002
A = 0.005
T = 1 - (R + A)
L = 3.25*25.4

F = 4*R/(1-R)**2
P0 = (T/(1-R))**2
B = 2*2*np.pi/c*L

P = P0*1./(1 + F * np.sin((B*nu + np.pi)/2)**2)
```

```python
dR = 1e-4
dR2 = 2e-5
Rquad = R - (nu - 150)*dR + (nu-150)**2*dR2
Rlin = R + (nu - 150)*dR

Rn = Rquad

Tn = 1 - (A + Rn)
Fn = 4*Rn/(1-Rn)**2
P0n = (Tn/(1-Rn))**2
Pn = P0n*1./(1 + Fn * np.sin((B*nu + np.pi)/2)**2)
```

```python
fig, ax = plt.subplots(figsize=(15,10))
ax.plot(nu, P, 'r', label=r'$R = {0:0.3f}$'.format(R))
ax.plot(nu, Pn, 'b', label=r'$R = {0:0.3f} - {1:0.4f}(\nu - 150)$'.format(R, dR))
ax.grid(which='both', axis='both')
ax.axis('tight')
# [140, 160, 0.6, 1.0]
ax.legend(loc='best')
ax.set_xticklabels(['{0:3.0f}'.format(t) for t in ax.get_xticks()]);
ax.set_yticklabels(['{0:0.4f}'.format(t) for t in ax.get_yticks()]);
```

```python
fig, ax = plt.subplots(figsize=(15,10))
ax.plot(nu, Rquad, 'r', label='quadratic')
ax.plot(nu, Rlin + (nu - 150)*dR, 'b', label='linear')
ax.axis('tight')
ax.legend(loc='best')
ax.set_xticklabels(['{0:3.0f}'.format(t) for t in ax.get_xticks()]);
ax.set_yticklabels(['{0:0.4f}'.format(t) for t in ax.get_yticks()]);
```
