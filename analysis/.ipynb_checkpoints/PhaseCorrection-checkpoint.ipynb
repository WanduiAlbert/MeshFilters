```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
from glob import glob
from astropy.constants import c
from scipy.fftpack import fft, fftfreq,ifft, rfftfreq
c = c.value*1e-6 #Speed of light in mm/ns
from scipy.signal import resample
from scipy.interpolate import interp1d
from scipy.optimize import curve_fit
```

```python
datadir = '../data/15-02 150GHz Ypol/'
files = glob(datadir + 'fts_150GHz_20160419_0*.txt') 
```

```python
index, time, encoder, signal = np.loadtxt(files[0], comments='#', skiprows=17, unpack=True)
encoder /= (c/2)
peak = encoder[np.where(signal == max(signal))[0]][0]
thresh = np.logical_and(encoder > (peak - 1.0), encoder < (peak + 1.0))
# fig, ax = plt.subplots(figsize=(15,10))
# ax.plot(encoder, signal, 'b');
# encoder = encoder[thresh]
# signal = signal[thresh]
```

```python
# Correction for the drift of the source
N = 400
mask = np.ones(N)
avg = np.convolve(signal, mask, 'same')/N
signal = signal[thresh]/avg[thresh] - 1
encoder = encoder[thresh]
# signal = signal[thresh]
fig, ax = plt.subplots(figsize=(15,10))
ax.plot(encoder, signal, 'b');
ax.axis('tight');
ax.set_xlabel('Path Length Difference [ps]');
```

```python
# Now lets try and correct for the phase
def quadratic(x, A, B, C):
    return A*x**2 + B*x + C

# First Find the right position of the origin of the interferogram
peak = np.where(signal == max(signal))[0][0]
y = signal[peak-1: peak+2]
x = encoder[peak-1:peak+2]
X = np.vstack([x**2, x, np.ones_like(x)]).T
A = np.linalg.solve(X, y)
actual_peak = -A[1]/(2*A[0])
sig_peak = quadratic(actual_peak, *list(A))
encoder -= actual_peak
encoder *=2 # Encoder now represents the path length difference
print actual_peak, sig_peak

# Next, we will resample the signal and path length difference as a 
```

```python
fig, ax = plt.subplots(figsize=(15,10))
ax.plot(encoder, signal, 'b');
ax.axis('tight');
```

```python
# Interpolate to an array of length 2**n with zero at the index 2**(n-1)-1. The array values go from -2 to 2
n = 12
k = 2.0
m = k*2**(1 - n)
encoder_new = m*np.arange(0, 2**n + 1)-2
# n = 12
# m = 4.0/(2**n - 1)
# encoder_new = m*np.arange(0, 2**n) - 2
encoder_new = encoder_new[:-1]
f = interp1d(encoder, signal, kind='cubic', bounds_error=False, fill_value=0.0)
signal_new = f(encoder_new)
```

```python
# n = 7
# m = 4.0/(2**n)
# x = m*np.arange(0, 2**n + 1) - 2
# print len(x)
# m*np.arange(2**n+1) - 2
print len(encoder_new)
```

```python
fig, ax = plt.subplots(figsize=(15,10))
ax.plot(encoder_new, signal_new, 'b');
ax.axis('tight');
```

```python
def get_odd(x):
    return 0.5*(x - np.roll(x[-1::-1], 1))

signal_odd = get_odd(signal_new)

# signal_new -= signal_odd

fig, ax = plt.subplots(figsize=(15,10))
ax.plot(encoder_new, signal_odd, 'b');
ax.axis('tight');

fig, ax = plt.subplots(figsize=(15,10))
ax.plot(encoder_new, signal_new, 'b');
ax.axis('tight');
```

```python
y = fft(signal_odd)
dx = (encoder_new[1] - encoder_new[0])
N = len(y)
k = fftfreq(N,dx)
# phi = -(1.09396927505)*k
# y *= np.exp(1j*phi)
k_thresh = np.logical_and(k >= 145, k <= 155)
fig, ax = plt.subplots(figsize=(15,10))
ax.plot(k[k_thresh], np.real(y[k_thresh]), 'r')
ax.plot(k[k_thresh], np.imag(y[k_thresh]), 'b');
```

```python
phi = np.unwrap(np.angle(y))
# phi = np.angle(y)
fig, ax = plt.subplots(figsize=(15,10))
ax.plot(k[k_thresh], phi[k_thresh] , 'r-');
ax.grid(which='both');
```

```python
def linear(x, m, b):
    return m*x + b
popt, pcov = curve_fit(linear, k[k_thresh], phi[k_thresh])
# ax.plot(k[k_thresh], linear(k[k_thresh], *popt), 'b+-')
y *= np.exp(-1j*linear(k, *popt));
fig, ax = plt.subplots(figsize=(15,10))
# ax.plot(k[k_thresh], phi[k_thresh]- linear(k[k_thresh], *popt), 'b+-');
ax.stem(k[k_thresh], phi[k_thresh] - linear(k[k_thresh], *popt) );
```

```python
k_thresh = np.logical_and(k >= 130, k <= 180)
fig, ax = plt.subplots(figsize=(15,10))
ax.plot(k[k_thresh], np.real(y[k_thresh]), 'r')
ax.plot(k[k_thresh], np.imag(y[k_thresh]), 'b');
```
